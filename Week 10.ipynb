```python
import matplotlib.pyplot as plt 
import numpy as np
from numpy.fft import fft, ifft
```

```python
# Exercise 1


N=64 
tau = 10.0
t = np.linspace(0, tau, N, endpoint=False) 

print(t)
#
f = 0.2 
x = np.cos(2*np.pi*f*t)
plt.plot(t, x)
plt.plot(t,x, 'o')

plt.title('Single cosine wave (against time) sampled at {} points'.format(N))
plt.xlabel('time (s)')
plt.ylabel('x(N)')
plt.grid()
plt.show()
```

```python
# Exercise 2

X = fft(x)/N 
plt.plot(np.abs(X),'o')
plt.title('FFT of above cosine wave sampled at '+str(N)+' points') 
plt.xlabel('point number')
plt.ylabel('$\mid X(N)\mid$')
plt.grid()
```

```python
# Exercise 3

Fs=1 / (t[1] - t[0])


def Spectrum1(x, Fs):
    '''
    One-sided amplitude spectrum of x(t).
    '''
    N = len(x)
    frq = (float(Fs)/N)*np.arange(0, N/2+1)
    X = fft(x)/N
    X = X[0:N//2+1]

    return frq , X

frq, X = Spectrum1(x, Fs)

plt.figure()
plt.plot(frq, np.abs(X))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Amplitude vs Frequency (Hz)')
plt.grid(True)
plt.show()

print("Why is the signal amplitude you find equal to 0.5 instead of 1?")
print("The signal amplitude is 0.5 instead of 1 because:   \n X = fft(x)/N \n X = X[0:N//2+1] \nThe fft is being divided by N")
```
